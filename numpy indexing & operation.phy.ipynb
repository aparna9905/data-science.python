```python
#example
expenses = np.array([
    [200, 340, 345, 654, 657], #food
    [100, 768, 564, 546, 876], #travel
    [300, 376, 765, 987, 435], #shopping
])
```

```python
print('total expenses: ',np.sum(expenses))  #these operations are very fast
print('category wise sum:',np.sum(expenses, axis =1))
print('day wise total: ', np.sum(expenses, axis =0))
print('highest expense day:', np.max(np.sum(expenses, axis=0)))
```

```python
#example2
marks = np.array([34,65,78,54,76])
```
