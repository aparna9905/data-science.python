```python
import random
import string

class bankaccount:
    
    def __init__(self, name, balance=0):
        self.name = name
        self.balance = balance
        self.transactions = []

        # generate username & password
        self.username = self.generate_username()
        self.password = self.generate_password()

        self.transactions.append(f"Name: {name}")
        self.transactions.append(f"Opening balance: {balance}")

        print("Welcome to our bank")
        print("Your credentials are:")
        print("Username:", self.username)
        print("Password:", self.password)

    # username generator
    def generate_username(self):
        return self.name.lower() + str(random.randint(100, 999))

    # strong password generator
    def generate_password(self):
        chars = string.ascii_letters + string.digits + string.punctuation
        return ''.join(random.choice(chars) for _ in range(10))

    # authentication check
    def authenticate(self):
        user = input("Enter username: ")
        pwd = input("Enter password: ")

        if user == self.username and pwd == self.password:
            return True
        else:
            print("Invalid credentials ❌")
            return False

    # deposit method
    def deposit(self, amount):
        if not self.authenticate():
            return

        if amount > 0:
            self.balance += amount
            self.transactions.append(
                f"Deposit: {amount}, Balance: {self.balance}"
            )
            print("Deposit successful ✅")
        else:
            print("Amount must be positive")

    # withdrawal method
    def withdraw(self, amount):
        if not self.authenticate():
            return

        if amount <= self.balance:
            self.balance -= amount
            self.transactions.append(
                f"Withdraw: {amount}, Balance: {self.balance}"
            )
            print("Withdrawal successful ✅")
        else:
            print("Insufficient balance ❌")

    # view balance
    def viewbalance(self):
        print(f"{self.name}, Balance: {self.balance}")

    # transaction history
    def transaction(self):
        print("Transaction History:")
        for t in self.transactions:
            print(t)
```

```python
acc.deposit(1000)
```
