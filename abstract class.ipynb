```python
from abc import ABC, abstractmethod
```

```python
class employee(ABC):
    @abstractmethod
    def calculate_salary(self):
        pass
    @abstractmethod
    def details(self):
        pass
```

```python
class employees1(employee):
    def __init__(self):
        self.name = input('enter  name')
        self.salary = int(input('enter salary'))
        self.pf = self.salary*0.20
        self.calculate_salary()# we call this function in constructor as if we make the object then this init will be called then if we call the details then 
    def calculate_salary(self):
        self.net_salary = self.salary + self.pf
        print(self.net_salary)
    def details(self):
        print(f"name of the employye is {self.name} and its salary is {self.net_salary}")
```

```python
user1=employees1()
```

```python
user1.calculate_salary()
```
